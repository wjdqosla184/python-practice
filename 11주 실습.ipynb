```python
import pandas as pd

# 데이터 생성
list3 = [
    ['Han_bit', 'Man', 20, 180],
    ['Han_la', 'Man', 21, 177],
    ['Ha_na', 'woman', 20, 160]
]
col_names = ['name', 'sex', 'age', 'height']

list3.append(['Jin_do', 'Man', 22, 175])
list3.append(['Ji_min', 'Woman', 23, 170])
list3.append(['Sung_mi', 'Woman', 21, 155])
list3.append(['Kim_nala', 'Woman', 24, 167])
list3.append(['Na_hyung', 'Man', 23, 170])

list3
```

```python
#복습 
# 1-1 iloc() 사용하여 행인덱스 2,3에 있는 나이, 키정보를 조회하라
df.iloc[2:4, 2:4] 

#iloc은 i가 붙은 만큼 index를 활용한 location 지정 방법으로 df.iloc[index_index, column_index] 방식이다.
#loc은 index를 활용하지 않고 직접 index 및 column 명을 통해 지정하는 방법으로
#df.loc[index_name, column_name]형식으로 나타낸다
#iloc은 데이터프레임의 위치를 기준으로 접근하고  loc은 데이터프레임의 인덱스와 컬럼명 기준으로 접근한다.
#iloc은 슬라이싱 마지막 값을 제외하고 loc은 끝값을 포함한다.
#iloc: 몇 번째 행/열 :정수 기반 접근
#loc: 어떤 이름의 행/열 : 라벨 기반 접근
```

```python
#1-2 loc() 사용하여 행인덱스 2,3에 있는 성별, 키 정보를 조회하라

df.loc[2:3, ['sex', 'height']]
```
