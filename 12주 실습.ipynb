```python
#복습 1-1
import pandas as pd

list1=list([['1020', '파이썬응용', '201호', 4],
            ['1007', '자료구조', '505호', 3],
            ['3002', '인공지능', '702호', 3]
           ])
col_names=['과목번호', '과목명', '강의실', '시수']
df=pd.DataFrame(list1, columns=col_names)
df
```

```python
#1-2 다음 데이터를 추가하라
#1003 c언어 205호 4
#2001 디지털영상처리 210호 4
df.loc[3] = ['1003', 'c언어', '205호', 4]
df.loc[4] = ['2001', '디지털영상처리', '210호', 4]
df
```

```python
#2-1 과목번호로 오름차순 정렬하라
df.sort_values(by=['과목번호'], ascending=False)
```

```python
#2-1 시수로 내림차순 정렬하라
df.sort_values(by=['시수'], ascending=False)
```

```python
#2-2 시수가 3인 교과목을 조회하라
df[(df['시수']==3)]
```

```python
#2-2 모든 과목의 시수를 1시간씩 감소하라
df.loc[0:4, '시수']=df.loc[0:4, '시수'] -1
df
```

```python
#2-3 인공지능 교과목의 정보를 조회하라
df[df['과목명'].str.contains('인공지능')]
```

```python
#실습
import pandas as pd
list2 = list([
    ['Han_bit', 'Man', 20, 180],
    ['Han_la', 'Man', 21, 177],
    ['Ha_na', 'Woman',20, 160]
])
col_names=['name', 'gender', 'age', 'height']
df=pd.DataFrame(list2, columns=col_names)
df
```

```python
df.set_index('name', inplace=True) #이름 열을 행 인덱스로
df
```

```python
#reset_index(inplace=True) 함수 인덱스를 기본 인덱스로 복구 시 사용
df.reset_index(inplace=True)
df
```

```python
#위에 데이터에서 age를 문자열로 나타내면 2020202020202020 처럼 반복 됨
df['bonus'] = df['age']*10000 # 나이에 10000을 곱한 만큼 보너스
df
```
