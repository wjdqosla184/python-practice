```python
import pandas as pd

# 파일 경로 설정
file_path = 'C:/Users/82102/Downloads/'

# 1. product.txt 파일 생성
product_data = {
    'product_name': ['Americano', 'Espresso', 'Capuccino', 'Ice tea'],
    'product_price': [3500, 4200, 4000, 4500]
}
product_df = pd.DataFrame(product_data)
product_file = file_path + 'product.txt'
product_df.to_csv(product_file, index=False)

# 2. option.txt 파일 생성
option_data = {
    'option_name': ['Short', 'Tall', 'Grande', 'Venti', 'add_pearlx1', 'add_pearlx2', 'add_pearlx3', 'add_shotx1', 'add_shotx2', 'add_shotx3', 'Takeout', 'Shop', 'ICE', 'HOT'],
    'description': ['220ml', '270ml', '350ml', '500ml', '', '', '', '', '', '', '', '', '', ''],
    'option_price': [0, 500, 1200, 1900, 500, 1000, 1500, 500, 1000, 1500, -1000, 0, 0, 0]
}
option_df = pd.DataFrame(option_data)
option_file = file_path + 'option.txt'
option_df.to_csv(option_file, index=False)

# 3. option_set.txt 파일 생성
option_set_data = {
    'option_set_name': ['size_0_3', 'size_0_3', 'size_0_3', 'size_0_3', 'size_0_1', 'shot_0_1', 'shot_0_2', 'shot_0_2', 'shot_0_3', 'shot_0_3', 'shot_0_3', 'pearl_0_1', 'pearl_0_2', 'pearl_0_2', 'pearl_0_3', 'pearl_0_3', 'pearl_0_3', 'ice_hot', 'ice_hot', 'takeout_shop', 'takeout_shop'],
    'option_name': ['Short', 'Tall', 'Grande', 'Venti', 'short', 'add_shotx1', 'add_shotx1', 'add_shotx2', 'add_shotx1', 'add_shotx2', 'add_shotx3', 'add_pearlx1', 'add_pearlx1', 'add_pearlx2', 'add_pearlx1', 'add_pearlx2', 'add_pearlx3', 'ICE', 'HOT', 'Takeout', 'Shop']
}
option_set_df = pd.DataFrame(option_set_data)
option_set_file = file_path + 'option_set.txt'
option_set_df.to_csv(option_set_file, index=False)

# 4. product_option_set.txt 파일 생성
product_option_set_data = {
    'product_name': ['Americano', 'Americano', 'Americano', 'Espresso', 'Capuccino', 'Capuccino', 'Capuccino', 'Ice tea', 'Ice tea'],
    'option_set_name': ['size_0_3', 'shot_0_3', 'ice_hot', 'shot_0_3', 'size_0_3', 'shot_0_3', 'ice_hot', 'size_0_3', 'shot_0_1']
}
product_option_set_df = pd.DataFrame(product_option_set_data)
product_option_set_file = file_path + 'product_option_set.txt'
product_option_set_df.to_csv(product_option_set_file, index=False)

# 5. 파일 읽기
product_df = pd.read_csv(product_file)
option_df = pd.read_csv(option_file)
option_set_df = pd.read_csv(option_set_file)
product_option_set_df = pd.read_csv(product_option_set_file)

# 각 제품별로 옵션을 추출하는 함수
def get_product_options(product_name):
    # 해당 제품의 option_set을 찾아옵니다.
    product_option_sets = product_option_set_df[product_option_set_df['product_name'] == product_name]
    option_sets = product_option_sets['option_set_name'].unique()

    # 각 option_set에 대한 옵션 목록을 만듭니다.
    options = []
    for option_set in option_sets:
        option_names = option_set_df[option_set_df['option_set_name'] == option_set]['option_name'].unique()
        for option_name in option_names:
            option = option_df[option_df['option_name'] == option_name].iloc[0]
            # description이 NaN이거나 빈 문자열일 경우 처리
            if pd.isna(option['description']) or option['description'] == '':  # 빈 description은 제외
                options.append({
                    'option_name': option['option_name'],
                    'description': '',  # description 빈칸
                    'option_price': option['option_price']
                })
            else:  # description이 있는 경우에는 괄호를 포함하여 출력
                options.append({
                    'option_name': option['option_name'],
                    'description': option['description'],
                    'option_price': option['option_price']
                })
    
    # 모든 옵션에 대해 Takeout을 추가하고 가격을 -1000으로 설정
    takeout_option = {
        'option_name': 'Takeout',
        'description': '',
        'option_price': -1000
    }
    options.append(takeout_option)
    
    return options

# 제품 가격 출력 및 옵션 선택 출력
def print_product_info():
    for _, product_row in product_df.iterrows():
        product_name = product_row['product_name']
        product_price = product_row['product_price']
        print()
        print(product_name, '-', product_price, '원')

        options = get_product_options(product_name)
        if options:
            print('가능한 옵션:')
            for option in options:
                if option['description']:
                    print('- ', option['option_name'], '(', option['description'], ') :', option['option_price'], '원')
                else:
                    print('- ', option['option_name'], ':', option['option_price'], '원')
        else:
            print('옵션이 없습니다.')

# 제품과 옵션 정보 출력
print_product_info()
```

```python
import pandas as pd

# 파일 경로 설정
file_path = 'C:/Users/82102/Downloads/'

# 1. product.txt 파일 생성
product_data = {
    'product_name': ['Americano', 'Espresso', 'Capuccino', 'Ice tea'],
    'product_price': [3500, 4200, 4000, 4500]
}
product_df = pd.DataFrame(product_data)

# 2. option.txt 파일 생성
option_data = {
    'option_name': ['Short', 'Tall', 'Grande', 'Venti', 'add_pearlx1', 'add_pearlx2', 'add_pearlx3', 'add_shotx1', 'add_shotx2', 'add_shotx3', 'Takeout', 'Shop', 'ICE', 'HOT'],
    'description': ['220ml', '270ml', '350ml', '500ml', '', '', '', '', '', '', '', '', '', ''],
    'option_price': [0, 500, 1200, 1900, 500, 1000, 1500, 500, 1000, 1500, -1000, 0, 0, 0]
}
option_df = pd.DataFrame(option_data)

# 3. option_set.txt 파일 생성
option_set_data = {
    'option_set_name': ['size_0_3', 'size_0_3', 'size_0_3', 'size_0_3', 'size_0_1', 'shot_0_1', 'shot_0_2', 'shot_0_2', 'shot_0_3', 'shot_0_3', 'shot_0_3', 'pearl_0_1', 'pearl_0_2', 'pearl_0_2', 'pearl_0_3', 'pearl_0_3', 'pearl_0_3', 'ice_hot', 'ice_hot', 'takeout_shop', 'takeout_shop'],
    'option_name': ['Short', 'Tall', 'Grande', 'Venti', 'short', 'add_shotx1', 'add_shotx1', 'add_shotx2', 'add_shotx1', 'add_shotx2', 'add_shotx3', 'add_pearlx1', 'add_pearlx1', 'add_pearlx2', 'add_pearlx1', 'add_pearlx2', 'add_pearlx3', 'ICE', 'HOT', 'Takeout', 'Shop']
}
option_set_df = pd.DataFrame(option_set_data)

# 4. product_option_set.txt 파일 생성
product_option_set_data = {
    'product_name': ['Americano', 'Americano', 'Americano', 'Espresso', 'Capuccino', 'Capuccino', 'Capuccino', 'Ice tea', 'Ice tea'],
    'option_set_name': ['size_0_3', 'shot_0_3', 'ice_hot', 'shot_0_3', 'size_0_3', 'shot_0_3', 'ice_hot', 'size_0_3', 'shot_0_1']
}
product_option_set_df = pd.DataFrame(product_option_set_data)

# 5. 파일 읽기 (이미 pandas DataFrame으로 읽혀져 있음)

# 각 제품별로 옵션을 추출하는 함수
def get_product_options(product_name):
    # 해당 제품의 option_set을 찾아옵니다.
    product_option_sets = product_option_set_df[product_option_set_df['product_name'] == product_name]
    option_sets = product_option_sets['option_set_name'].unique()

    # 각 option_set에 대한 옵션 목록을 만듭니다.
    options = []
    for option_set in option_sets:
        option_names = option_set_df[option_set_df['option_set_name'] == option_set]['option_name'].unique()
        for option_name in option_names:
            option = option_df[option_df['option_name'] == option_name].iloc[0]
            # description이 NaN이거나 빈 문자열일 경우 처리
            if pd.isna(option['description']) or option['description'] == '':  # 빈 description은 제외
                options.append({
                    'option_name': option['option_name'],
                    'description': '',  # description 빈칸
                    'option_price': option['option_price']
                })
            else:  # description이 있는 경우에는 괄호를 포함하여 출력
                options.append({
                    'option_name': option['option_name'],
                    'description': option['description'],
                    'option_price': option['option_price']
                })
    
    # 모든 옵션에 대해 Takeout을 추가하고 가격을 -1000으로 설정
    takeout_option = {
        'option_name': 'Takeout',
        'description': '',
        'option_price': -1000
    }
    options.append(takeout_option)
    
    return options

# 제품 가격 출력 및 옵션 선택 출력
def print_product_info():
    product_info_list = []  # 모든 정보들을 담을 리스트
    for _, product_row in product_df.iterrows():
        product_name = product_row['product_name']
        product_price = product_row['product_price']
        
        product_info = {
            'product_name': product_name,
            'product_price': product_price
        }
        
        options = get_product_options(product_name)
        option_info = []
        for option in options:
            option_info.append({
                'option_name': option['option_name'],
                'description': option['description'],
                'option_price': option['option_price']
            })
        
        product_info['options'] = option_info  # 해당 제품의 옵션 추가
        product_info_list.append(product_info)
    
    return product_info_list

# 결과 출력 (제품과 옵션 모두 출력)
final_info = print_product_info()

# 포인트 적립 열 추가 (가격의 10% 포인트)
for product_info in final_info:
    product_info['point_earned'] = product_info['product_price'] * 0.1  # 제품 가격의 10%

# 전체 결과 출력
for product_info in final_info:
    print(f"제품명: {product_info['product_name']}, 가격: {product_info['product_price']}원, 포인트: {product_info['point_earned']}원")
    print("옵션:")
    for option in product_info['options']:
        if option['description']:
            print(f"- {option['option_name']} ({option['description']}) : {option['option_price']}원")
        else:
            print(f"- {option['option_name']} : {option['option_price']}원")
    print("-" * 50)
```
